```python
import corner
import emcee
import matplotlib.pyplot as plt
import numpy as np
import rebound

%matplotlib inline
```

# Define constants and an array of observing times

```python
AUperTwoPiYearToMetersPerSecond = 4743.72 * 2 * np.pi
DaysInYear = 365.2425
DaysToTwoPiYears = 2.*np.pi/DaysInYear
MsunToMJupiter = 1047.56

times = np.array(sorted(np.random.uniform(0., 1000., 150)))
```

# Specify the stellar mass, planetary system, a single offset, and a single jitter.

```python
### mass of the central star in [solar masses]
mstar = 1.

### orbital periods in [days]
P = [500., 800.]

### semi-amplitudes in [m/s]
m = [0.0004, 0.0008]

### eccentricities
e = [0.1, 0.07] 

### argument of pericenter in [radians]
w = [np.e, np.pi] 

### time of periastron passage in [days]
T = [1., 2.]

### zero-point offset(s) in [m/s]
offset = [-3.]

### jitter in [m/s]
jitter = [2.]

assert len(P)==len(m)==len(e)==len(w)==len(T)

npl = len(P)

theta = []
for p in range(npl):
    theta.append(P[p])
    theta.append(m[p])
    theta.append(e[p])
    theta.append(w[p])
    theta.append(T[p])
    
theta += offset
theta += jitter
```

# Function that returns RVs for an input planetary system

```python
def rvs_from_nbody_model(mstar=1., P=[365.], m=[2.], e=[0.04], w=[0.], T=[0.], times=times, npl=npl):    
    times_2pi = times * DaysToTwoPiYears
    
    sim = rebound.Simulation()
    sim.add(m=mstar)
    
    for p in range(npl):
        sim.add(P=P[p]*DaysToTwoPiYears, m=m[p], e=e[p], omega=w[p], T=T[p]*DaysToTwoPiYears)
    sim.move_to_com()

    rvs = np.zeros(len(times))

    for i, t in enumerate(times):
        sim.integrate(t)
        rvs[i] = -sim.particles[0].vy * AUperTwoPiYearToMetersPerSecond
        
    return rvs
```

```python
rvs_real = rvs_from_nbody_model(mstar=mstar, P=P, m=m, e=e, w=w, T=T, times=times) + offset[0]
```

```python
sigmas = np.random.randn(len(rvs_real))*0.1 + 1

rvs = rvs_real + np.random.randn(len(rvs_real)) * sigmas + np.random.randn(len(rvs_real)) * jitter
```

```python
plt.errorbar(times, rvs_real, yerr=sigmas, fmt='o')
plt.xlabel("Time [days]", fontsize=16)
plt.ylabel("RV [m/s]", fontsize=16)
plt.show()
```

# Functions for loglikelihood, logprior, and logposterior

```python
def lnlike(theta, obs, mstar=1, npl=npl):
    '''
    The log-likelihood function. Observational model assume RVs have independent Gaussian error bars.
    '''
    P, m, e, w, T = np.reshape(theta[0:5*npl], (2,5)).T
    offsets = theta[5*npl:-1]
    jitter = theta[-1]
    
    times, rvs, sigmas = obs
    
    model = rvs_from_nbody_model(mstar=mstar, P=P, m=m, e=e, w=w, T=T, times=times, npl=npl) + offsets[0]
    inv_sigma2 = 1.0/(sigmas*sigmas + jitter*jitter)
    return -0.5*( np.sum( (rvs - model)*(rvs - model)*inv_sigma2 + np.log((2.*np.pi)/inv_sigma2) ) )


def lnprior(theta, mstar=1, npl=npl):
    '''
    The log-prior function. Excluding prior in offset since we aren't comparing models with different offsets.
        
         Parameters
         ----------
         theta : 1-d list of Keplerian orbital parameters (p, K, e, w, M)
                 for every modeled planet
    '''
    P, m, e, w, T = np.reshape(theta[0:5*npl], (2,5)).T
    offsets = theta[5*npl:-1]
    jitter = theta[-1]

    pmax, mmax, cmax, jmax = 999999., 999999., 999999., 999999.
    
    lnp = 0.

    if all( 0. < i < pmax for i in P) and all( 0. < i < mmax for i in m) and all( 0. <= i < 1 for i in e) \
    and all(-cmax < i < cmax for i in offsets) and 0. < jitter < jmax:
        lp = 1.
    else:
        return -np.inf
    
    return lnp

def lnpost(theta, obs, mstar=1, npl=npl):
    '''
    The log-posterior function. Sum of log-likelihood and log-prior.
        
        Parameters
        ----------
        theta : 1-d list of Keplerian orbital parameters (p, K, e, w, M)
                for every modeled planet
        obs   : observations object with radial velocity measurement (.rv)
                and uncertainty (.err) attributes
    '''
    lp = lnprior(theta, mstar, npl)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta, obs, mstar, npl)
```

# Testing an evaluation of loglikelihood

```python
lnlike(theta, [times, rvs, sigmas], mstar=mstar, npl=npl)
```

# Find parameter values that maximize likelihood

```python
import scipy.optimize as op
nll = lambda *args: -lnlike(*args)

def bounds(theta, npl):
    '''
    Bounds function. Ensures gravitationally bound orbits (0 < e < 1) and some non-negative quantities.
        
        Parameters
        ----------
        theta : 1-d list of Keplerian orbital parameters (p, K, e, w, M)
                for every modeled planet
    '''
    bnds = ()
    for i in range(npl):
        bnds = bnds + ((0., None), (0., None), (0.,0.99999), (-4.*np.pi, 4.*np.pi), (-4.*np.pi, 4.*np.pi))
        
    bnds = bnds + ((-100.,100.),) + ((-100.,100.),)
    return bnds
```

```python
result = op.minimize(nll, theta, args=([times, rvs, sigmas], mstar), bounds=bounds(theta, npl))
```

```python
result['x']
```

# Start the MCMC

```python
ndim = len(result['x'])
nwalkers = 2*ndim
pos = [result["x"] + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
```

```python
import emcee
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnpost, args=([times, rvs, sigmas], mstar, npl))
```

```python
nsteps = 10000
sampler.run_mcmc(pos, nsteps, thin=10);
```

# Format emcee output to array that can be read into a cornerplot

```python
samples = sampler.chain[:, 500:, :].reshape((-1, ndim))
```

```python
fig = corner.corner(samples, labels=["$P1$", "$K1$", "$e1$", "$w1$", "$T1$", "$P2$", "$K2", "$e1$", "$w1$", "$T1$", "$C$", "$J$"],
                   truths=theta)
#fig.savefig("triangle.png")
```
